```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
x=np.arange(0,10)
y=np.arange(11,21)
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d graph')
plt.savefig('test.png')
```

```python
y=x*x
plt.plot(x,y)
```

```python
y=x*x
plt.plot(x,y,'r--')
```

```python
y=x*x
plt.plot(x,y,'r*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'m')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib demo')
plt.plot(x,y,'r*-')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```
